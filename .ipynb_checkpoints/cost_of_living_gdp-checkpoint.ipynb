## Cost of Living vs. GDP (by Country)
The proposed hypothesis for this notebook is: "Cost of Living increases with the increase of a Nations Gross Domestic Product". 
To test this hypothesis the following notebook will look to explore the relation between these two common performance metrics of a range of countries.

### Import libraries and prepare the general dataset
Import the appropriate libraries that will be used for analysis throughout the notebook. Additionally, perpare the dataset for analysis.
This dataset was acquired from the OECD Statistics data website, and comprises the GDP of the OECD member nations over the period 2016 to present.

```python
# Initial imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load current .csv dataset for GDP data
data = "datasets/GDP_Selected_2000_2020_OECD.csv" # Represents most recent data for each country, some countries report data more regularly

# Load csv data into a DataFrame
raw_df = pd.read_csv(data)

raw_df.head()
```

```python
# Sort the GDP data by Year.
df = raw_df.copy().sort_values(by='TIME', ascending=False)
df.head()
```

### Visualize GDP Data
The data was then plotted for a specific year to get a visual of the data distribution.

```python
# Plot the GDP data for each country code.
temp_df = df.loc[df['TIME'] == 2019]
temp_df = temp_df.sort_values(by='Value', ascending=False)
temp_df[['LOCATION','Value']].plot(kind='bar', x='LOCATION', figsize=(12,6), title='GDP Value (USD/Capita) by Country Code for 2019')
plt.show()
```

### Get Country names from Country Code
Due to the format of the GDP data that was aqcuired the label of each country in the dataset was a 3 letter country code, rather than the country's name itself.
To allow the data to be more human readable, the data would need appropriate labelling to identify the country by name.
Through some additional processing these country codes could be matched with their corresponding country name.

```python
# Fetch country code from website.
codes_df = pd.read_html('https://laendercode.net/en/3-letter-list.html')[0]
codes_df.head()
```

```python
# Join the sorted data on the Country code, to get appropriate country name
merged_df = df.merge(codes_df, left_on=['LOCATION'], right_on=['ISO 3166 ALPHA-3'])
merged_df.head()
```

```python
# Plot data with country names for a single year
temp_df = merged_df.loc[df['TIME'] == 2019]
temp_df = temp_df.sort_values(by='Value', ascending=False)
temp_df[['Country','Value']].plot(kind='bar', x='Country', figsize=(12,6), title='GDP Value (USD/Capita) by Country for 2019')
plt.show()
```

After this processing, it can be seen that the GDP Value data was now more human readable with the Country names present.

The next step in the analysis was to combine this data with Cost of Living data and identify any possible relations between the datasets.

## Acquire the Cost of Living Index data
The cost of living index data was available through the Numbeo website (https://www.numbeo.com/cost-of-living/rankings_by_country.jsp) for academic use, which was ideal for this project.

The data gathered by Numbeo for the cost of living index, was a relative index for the price of consumer goods (including rent) compared to New York City. This relative index acts as a percentage comparison between New York (100%) and the reported nations.

```python
# Load table data from Numbeo for the Cost of Living Index data for the year 2020

# Years to process
years = list(range(2009,2021)) # based on the availability of the Numbeo data
col_df = pd.DataFrame()

for year in years:
    url = 'https://www.numbeo.com/cost-of-living/rankings_by_country.jsp?title='+str(year)+'&displayColumn=2' # Includes rent
    annual_df = pd.read_html(url)[1]
    annual_df['Year']=year
    col_df = col_df.append(annual_df)

col_df = col_df.drop(columns='Rank')
col_df
```

### Combine Datasets
With the GDP and Cost of Living data now available and processed appropriately, these datasets could be combined for further analysis.

```python
# Merge the Cost of Living data with GDP data for each common Country and Year
merged_df = merged_df.merge(col_df, left_on=['Country','TIME'], right_on=['Country','Year'])
merged_df
```

## Plot the Merged Data for one Year (2019)
To visualize the now merged data, it was appropriate to plot the two metrics against one another.
From this plot possible new analysis could be pursued.

```python
# Create a figure for the combined plot
fig,ax = plt.subplots(figsize=(12,6))

ex_plot_df = merged_df.loc[merged_df['TIME']==2019].sort_values(by='Value', ascending=False)

# Plot the GDP data on the first axis
ax.plot(ex_plot_df['Country'], ex_plot_df['Value'], color="darkorange", marker="o")
ax.set_xlabel('Country', fontsize=14)
ax.set_ylabel("GDP (USD/Capita)",color="darkorange",fontsize=14)
ax.set_ylim(ymin=0)

# Create a twin object from the first axis, to allow two different y-axis on the same plot
ax2=ax.twinx()

# Plot the Cost of Living data on the second axis
ax2.plot(ex_plot_df['Country'], ex_plot_df['Cost of Living Plus Rent Index'], color="cornflowerblue", marker="o")
ax2.set_ylabel("Cost of Living Index (% vs. NYC)", color="cornflowerblue", fontsize=14)
ax2.set_ylim(ymin=0)
ax.set_title('GDP vs. Cost of Living (by Country) for 2019')

# Ensure the country labels are rotated to be readable
ax.tick_params(axis='x', labelrotation = 90)

plt.show()
```

## GDP and Cost of Living Correlation
From the plot comparison there is some visual evidence that a relation exists between the GDP of a nation and its Cost of Living.
Further analysis would help clarify these initial impressions, which could be performed through the Pearson's Correlation coefficient.
Due to the variability in the data, the Spearman Rank Correlation would also be performed.

```python
# Pearson's Correlation Coefficient and Spearman's Rank Correlation
pearson = merged_df.loc[merged_df['TIME']==2019]['Value'].corr(merged_df['Cost of Living Plus Rent Index'])
spearman = merged_df.loc[merged_df['TIME']==2019]['Value'].corr(merged_df['Cost of Living Plus Rent Index'], method='spearman')

print('Correlation results between GDP Value and Cost of Living Index:')
print(f'Pearson\'s Correlation Coefficient:\t {pearson}')
print(f'Spearman\'s Rank Correlation:\t\t {spearman}')
```

### Correlation Results for 2019
Since both of these Correlation metrics approach the max correlation range of 1, this indicated that there was a strong relation between the variables 'GDP Value' and 'Cost of Living'.
Statistically, a correlation value greater than 0.8 is significant, and indicates that these values are closely connected. Here the results are both near Positive 1, which implies that as GDP increases for a given country, so to would its Cost of Living be expected to increase by a similar proportion.

### Further Correlation Analysis
Due to the strong correlation identified for the selected year of 2019, further analysis was performed to evaluate the correlation metrics for each year in the dataset.

```python
# Repeat the above to compute Correlation over the range of years
cor_list = []

# Compute the correlation values for each year in the set.
for year in years:
    temp_list = []
    temp_list.append(year)
    temp_list.append(merged_df.loc[merged_df['TIME']==year]['Value'].corr(merged_df['Cost of Living Plus Rent Index']))
    temp_list.append(merged_df.loc[merged_df['TIME']==year]['Value'].corr(merged_df['Cost of Living Plus Rent Index'], method='spearman'))
    cor_list.append(temp_list)

correlation_df = pd.DataFrame(data=cor_list, columns=['Year','Pearson','Spearman'])
correlation_df.head()
```

### Plotting the Correlation metrics over Time
With the correlation metrics computed for all the available years of data, the results could be plotted to visually identify any trends.

```python
# Plot correlation results over time
fig,ax = plt.subplots(figsize=(12,6))

# Plot the Pearson correlation data
ax.plot(correlation_df['Year'], correlation_df['Pearson'], label = 'Pearson')
ax.set_xlabel('Year', fontsize=14)
ax.set_ylabel("Correlation Value", fontsize=14)
ax.set_ylim(ymin=0, ymax=1)

# Plot the Spearman rank correlation data
ax.plot(correlation_df['Year'], correlation_df['Spearman'], label = "Spearman")
ax.set_title('Correlation of GDP vs Cost of Living Index over Time')

ax.grid(which='major', alpha=0.5)
plt.hlines(0.8, 2009, 2020, colors='indianred', linestyles='dashed', label='Significance Threshold')
plt.legend(loc='center right')
plt.show()
```

## Correlation over Time Analysis
As before, both of the Correlation metrics largely lay above the significance threshold of correlation value 0.8, for each year analysed. This indicates that for the data that was available, there has been a strong correlation between the GDP of a country and its Cost of Living. The strength of this correlation has remained largely unchanged over the data range analysed, which implies that these economic metrics are closely linked.


## Hypothesis Result
The analysis performed here sought to challenge the Hypothesis: "Cost of Living increases with the increase of a Nations Gross Domestic Product". The results have shown a close connection between these two economic metrics of a nation, and validate that this proposed hypothesis was acceptable.
