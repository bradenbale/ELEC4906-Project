```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('world-gdp-growth-rate.csv') #read the GDP file
df = pd.DataFrame(data) #create a dataframe
data2 =  pd.read_csv('world-unemployment-rate.csv')
df2 = pd.DataFrame(data2) #create a dataframe
```

```python
years = np.linspace(1991,2019,29).astype('float')
array = []
slope = []
intercept = []

array = []
array.append(df.iloc[0:29,1])
s, i = np.polyfit(years, array[0].astype('float'), 1)
slope.append(s) 
intercept.append(i) 

array.append(df2.iloc[0:29,1])
s, i = np.polyfit(years, array[1].astype('float'), 1)
slope.append(s) 
intercept.append(i) 
```

```python
names = ['GDP','Unemployment rate']

for data,colour,name in zip(array,['red','blue'],names):
  i = names.index(name)
  line = [slope[i]*year+intercept[i] for year in years]
  plt.plot(years,array[i],color=colour,marker='o',linestyle='none',label=name)
  plt.plot(years,line,color=colour,linestyle='solid',label='Slope: '+round(slope[i],3).astype('str'))
plt.legend()
plt.xlabel('Year (1991 - 2019)')
plt.title('Scatter Plot (GDP & Unemployment Rate)')
plt.show()
```
