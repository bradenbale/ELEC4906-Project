```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure


data = pd.read_csv('world-gdp-growth-rate.csv') #read the GDP file
df = pd.DataFrame(data) #create a dataframe
data2 =  pd.read_csv('world-unemployment-rate.csv')
df2 = pd.DataFrame(data2) #create a dataframe
data3 = pd.read_csv('world-gdp-gross-domestic-product.csv')
df3 = pd.DataFrame(data3)
```

```python
years = np.linspace(1991,2019,29).astype('float')
array = []
slope = []
intercept = []

array = []
array.append(df.iloc[0:29,1])
s, i = np.polyfit(years, array[0].astype('float'), 1)
slope.append(s) 
intercept.append(i) 

array.append(df2.iloc[0:29,1])
s, i = np.polyfit(years, array[1].astype('float'), 1)
slope.append(s) 
intercept.append(i) 
```

```python
names = ['GDP','Unemployment rate']

figure(figsize=(8, 5), dpi=80)

for data,colour,name in zip(array,['red','blue'],names):
  i = names.index(name)
  line = [slope[i]*year+intercept[i] for year in years]
  plt.plot(years,array[i],color=colour,marker='o',linestyle='none',label=name)
  plt.plot(years,line,color=colour,linestyle='solid',label='Slope: '+round(slope[i],5).astype('str'))
plt.legend()
plt.xlabel('Year')
plt.ylabel('Percentage (%)')
plt.title('Scatter Plot (GDP & Unemployment Rate)')
axes = plt.gca()
axes.set_ylim([-10,10])
plt.show()
```

```python
names = ['GDP Growth rate','Unemployment rate']

figure(figsize=(8, 5), dpi=80)

for data,colour,name in zip(array,['red','blue'],names):
  i = names.index(name)
  plt.plot(years,array[i],color=colour,marker='o',linestyle='solid',label=name)
plt.legend()
plt.xlabel('Year')
plt.ylabel('Percentage (%)')
plt.title('Scatter Plot (GDP growth rate & Unemployment Rate)')
axes = plt.gca()
axes.set_ylim([-5,10])
plt.grid()
plt.show()
```

```python
# Pearson's Correlation Coefficient
pearson = array[0].corr(array[1])
spearman = array[0].corr(array[1], method='spearman')
print(f'Pearson\'s Correlation Coefficient between GDP growth rate and Unemployment Rate: {pearson}')
print(f'Spearman\'s Rank Correlation between GDP growth rate and Unemployment rate {spearman}')
```
